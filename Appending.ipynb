```python
import os
import numpy as np
import mne
import pyedflib
import pandas
import matplotlib.pyplot as plt
from scipy import signal
import math
```

```python
# Define the path to the directory containing the .edf files
data_dir = "D:\\Minor Project\\eeg-motor-movementimagery-dataset-1.0.0\\Training Data\\input"

# Get a list of all .edf files in the directory
edf_files = [f for f in os.listdir(data_dir) if f.endswith('.edf')]

# Create an empty list to store the raw data objects
raw_list = []

# Loop through each .edf file and create a raw data object
for edf_file in edf_files:
    edf_path = os.path.join(data_dir, edf_file)
    raw = mne.io.read_raw_edf(edf_path)
    raw_list.append(raw)

# Concatenate the raw data objects into a single file
raw_concatenated = mne.concatenate_raws(raw_list)

# Save the concatenated raw data to a new file
raw_concatenated.save("D:\\Minor Project\\eeg-motor-movementimagery-dataset-1.0.0\\Training Data\\output2\\output2_file.fif", overwrite=True)

# Get the events from the concatenated raw data
events = mne.events_from_annotations(raw_concatenated)


# Save the events to a separate file
mne.write_events("D:\\Minor Project\\eeg-motor-movementimagery-dataset-1.0.0\\Training Data\\event2\\event2_file.tsv", events,overwrite=True)
```
